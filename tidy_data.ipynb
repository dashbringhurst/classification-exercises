Attendance Data

Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with something like this:


```python
import numpy as np
import pandas as pd
import acquire
```

```python
tidy = acquire.get_tidy_data_attendance()
```

```python
tidy.info()
```

```python
tidy = tidy.rename(columns={'Unnamed: 0':'name'})
```

```python
tidy_melt = tidy.melt(id_vars='name', var_name='date', value_name='attendance')
tidy_melt.attendance = tidy_melt.attendance.str.replace('P','1.0').str.replace('A', '0').str.replace('T','.9').str.replace('H','.5').astype(float)
```

```python
tidy_melt.groupby('name').attendance.mean()
```

Cake Recipes

Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

Tidy the data as necessary.

Which recipe, on average, is the best? recipe b

Which oven temperature, on average, produces the best results? 275

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake = acquire.get_tidy_data_cake()
```

```python
cake.info()
```

```python
cake.head()
```

```python
cake_melt = cake.melt(id_vars=['recipe:position'], var_name= 'temp', value_name='rating')
```

```python
cake_melt[['recipe', 'position']] = cake_melt['recipe:position'].str.split(':', expand = True)
cake_melt
```

```python
cake_melt = cake_melt.pivot_table(index=['recipe','position','temp']).reset_index()
```

```python
cake_melt.groupby('recipe').rating.mean()
```

```python
cake_melt.groupby('temp').rating.mean()
```

```python
cake_melt[cake_melt.rating == cake_melt.rating.max()]
```
